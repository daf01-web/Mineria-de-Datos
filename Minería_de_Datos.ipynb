#### Conceptos

Se refiere al proceso o técnicas de extraer información útil a partir de una acumulación de datos.

Su propósito principal es examinar grandes data sets para identificar tendencias, patrones y relaciones a fin de dar soporte a la toma de decisiones y planificación fundamentadas.

#### Ciencia de Datos

Disciplina que combina técnicas y recursos de estadística, matemáticas, ciencia de la computación y la inteligencia artificial con la finalidad de obtener información de interés.

Es un campo científico y académico centrado en el análisis de datos masivos (big data) para la extracción de información útil o valiosa, hace visibles conocimientos que se encuentran ocultos o implícitos en bases de datos.
